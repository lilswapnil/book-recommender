# Jupyter Notebook Setup Guide

This notebook will guide you through the steps to set up and launch Jupyter Notebook in a new environment.

## 1. Check Python Installation

First, check if Python is installed on your system. Open a terminal and run one of the following commands:

```python
# Check Python version
!python3 --version || python --version
```

## 2. Install pip

If pip is not installed, you can install it by downloading the official get-pip.py script or using your system's package manager. For most systems, you can try:

```python
# Try installing pip (if not already installed)
!python3 -m ensurepip --upgrade || curl https://bootstrap.pypa.io/get-pip.py -o get-pip.py && python3 get-pip.py
```

## 3. Install Jupyter Notebook

Once pip is installed, you can install Jupyter Notebook using the following command:

```python
# Install Jupyter Notebook
!python3 -m pip install notebook
```

## 4. Launch Jupyter Notebook

After installation, you can launch Jupyter Notebook by running the following command in your terminal:

```python
# Launch Jupyter Notebook
!jupyter notebook
```
