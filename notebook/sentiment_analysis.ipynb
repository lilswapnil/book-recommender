```python
emotions = ['anger', 'disgust', 'fear', 'joy', 'neutral', 'sadness', 'surprise']

def calculate_dependent_emotions_score(predictions):
    """
    predictions: output of classifier(list_of_sentences) when top_k=None
    returns: dict[label] = max score for that label across all sentences
    """
    max_scores = {e: 0.0 for e in emotions}

    for sent_pred in predictions:           # one sentence
        for item in sent_pred:              # each emotion label for that sentence
            label = item["label"]
            score = float(item["score"])    # <- no np.float64
            if label in max_scores and score > max_scores[label]:
                max_scores[label] = score

    return max_scores
```

```python
isbns = []
emotion_score = {label: [] for label in emotions}

for i in range(10):
    isbns.append(loaded_books_df["isbn"].iloc[i])
    sentences = [s.strip() for s in str(loaded_books_df["desc"].iloc[i]).split(".") if s.strip()]

    predictions = classifier(sentences, top_k=None)  # IMPORTANT
    max_scores = calculate_dependent_emotions_score(predictions)

    for label in emotions:
        emotion_score[label].append(max_scores[label])
```

## Chunking helper (token-based, not sentence-based)

```python
from transformers import AutoTokenizer
import numpy as np

model_name = "j-hartmann/emotion-english-distilroberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)

emotions = ['anger','disgust','fear','joy','neutral','sadness','surprise']

def chunk_text(text, max_tokens=512):
    # returns list of chunk strings, each <= max_tokens
    if not text:
        return [""]
    ids = tokenizer.encode(text, add_special_tokens=False)
    chunks = []
    for i in range(0, len(ids), max_tokens):
        chunk_ids = ids[i:i+max_tokens]
        chunks.append(tokenizer.decode(chunk_ids, skip_special_tokens=True))
    return chunks
```

## Aggregate emotion scores across chunks (no truncation)

```python
def aggregate_chunk_predictions(chunk_preds, agg="mean"):
    # chunk_preds: list of per-chunk predictions (each is list of {label, score})
    # returns dict[label] -> aggregated score
    per_label = {e: [] for e in emotions}
    for pred in chunk_preds:
        m = {d["label"]: float(d["score"]) for d in pred}
        for e in emotions:
            per_label[e].append(m.get(e, 0.0))

    if agg == "max":
        return {e: float(np.max(per_label[e])) for e in emotions}
    else:  # mean
        return {e: float(np.mean(per_label[e])) for e in emotions}
```

## Batch over chunks (fast) and map back to books

```python
from tqdm import tqdm
import pandas as pd

texts = loaded_books_df["desc"].fillna("").astype(str).tolist()
isbns = loaded_books_df["isbn"].tolist()

batch_size = 32

all_scores = {e: [] for e in emotions}

for text in tqdm(texts):
    chunks = chunk_text(text, max_tokens=512)

    # run classifier on all chunks in batches
    chunk_results = []
    for start in range(0, len(chunks), batch_size):
        chunk_batch = chunks[start:start+batch_size]
        chunk_results.extend(classifier(chunk_batch, top_k=None, truncation=True, max_length=512))

    scores = aggregate_chunk_predictions(chunk_results, agg="mean")  # or "max"

    for e in emotions:
        all_scores[e].append(scores[e])

# attach to df
for e in emotions:
    loaded_books_df[e] = all_scores[e]
```

```python
emotions_df = pd.DataFrame(emotion_score)
emotions_df['isbn'] = isbns
```
